```python
def sql_to_df(query):
    con = sqlite3.connect('sakila.db')
    df = pd.read_sql(query, con)
    return df
```

```python
sql_to_df("SELECT first_name,last_name FROM customer")
```

```python
sql_to_df("SELECT DISTINCT(country_id) FROM city")
```

```python
sql_to_df("SELECT * FROM customer WHERE store_id = 1")
```

```python
sql_to_df("SELECT * FROM film WHERE release_year = 2006 and rating = 'R' ")
```

# count:

```python
query = """ SELECT COUNT(customer_id) FROM customer"""
sql_to_df(query)
```

# wildcard

```python
sql_to_df(""" SELECT * FROM customer WHERE first_name LIKE 'm%' ; """)
```

```python
sql_to_df(""" SELECT * FROM customer WHERE first_name LIKE 'M_R%' ; """)
```

# ORDER BY

```python
sql_to_df(""" SELECT * FROM customer ORDER BY last_name ASC;""")
```

# GROUP BY

```python
sql_to_df(""" SELECT store_id, COUNT(customer_id) FROM customer GROUP BY store_id""")
```
