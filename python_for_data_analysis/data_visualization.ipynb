# boxplots and violin plots

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data1 = stats.norm(0,5).rvs(100)


data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])


sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot([data1,data2],inner='stick')
```

# Regression plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(tips,x='total_bill',y='tip')
```

```python
sns.lmplot(tips,x='total_bill',y='tip',
            scatter_kws={'marker':'o','color':'indianred'},
            line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot(tips,x='total_bill',y='tip',order=3,
            scatter_kws={'marker':'o','color':'indianred'},
            line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot(tips,x='size',y='tip')
```

```python
import numpy as np
sns.lmplot(tips,x='size',y='tip',x_estimator=np.mean,fit_reg=False)
```

```python
sns.lmplot(tips,x='size',y='tip',x_estimator=np.min,fit_reg=False)
```

```python
sns.lmplot(tips,x='size',y='tip',fit_reg=False)
```

```python
tips
```

```python
sns.lmplot(tips,x='size',y='total_bill',x_estimator=np.mean,hue='day',)
```

```python
sns.violinplot(tips,x='size',y='total_bill')
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)
sns.regplot(tips,y='total_bill',x='size',ax=axis1)
sns.violinplot(tips,x='size',y='total_bill',ax=axis2)
```

# Heatmap

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights = flights.pivot(index='month',columns='year',values='passengers')
```

```python
flights.head()
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot=True, fmt='d')
```

```python
flights.head()
```

### configurando o 0 do range de temperatura:

```python
sns.heatmap(flights,center=flights.loc['Jan',1955]) 
```

### Clustermaps (cluster similar values)

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights,col_cluster=False)
```

```python
sns.clustermap(flights,z_score=1)
```
