```python
customer_who_bought = customer[product['Status'] == 'Sold']
```

```python
frequency_table = customer_who_bought[["Country","Customer ID"]].groupby(by="Country").count()
```

```python
frequency_table.rename({"Customer ID":'number_of_clients'},axis=1,inplace=True)
```

```python
frequency_table['relative'] = 100*frequency_table['number_of_clients']/len(customer_who_bought)
```

```python
frequency_table
```

```python
frequency_table.sort_values(by='number_of_clients',ascending=False,inplace=True)
```

```python
frequency_table['cumsum'] = frequency_table.relative.cumsum()
```

```python
frequency_table
```

```python
import matplotlib.pyplot as plt
```

```python
fig,ax =  plt.subplots()
ax.bar(x=frequency_table.index,height=frequency_table['number_of_clients'])
ax2 = ax.twinx()
#ax2.plot(df['city'],df['prob'],c='red')
ax2.set_ylim(0,100)
ax2.plot(frequency_table.index,frequency_table['cumsum'],c='red')
```
